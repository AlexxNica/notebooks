## Model Analysis

Analyze the model by looking at the summary events the training program outputs. Let's look at the events the local run generated in previous notebook, but events generated by cloud run in GCS works too.<br>

First list events.

```python
%ml summary --dir /content/datalab/ml/census/model/summaries
```

Plot events.

```python
%ml summary --dir /content/datalab/ml/census/model/summaries --name loss relative_error
```

Note that you can plot events in the same graph from multiple directories (--dir support multiple values, just like --name). This is useful if you want to compare two model's training progress. <br>
