## Model Analysis

Analyze the model by looking at the summary events the training program outputs. Let's look at the events the two local run generates in previous notebook, but events generated by cloud run works too.

Using "%ml summary" command to view events. First list events in a given dir.

```python
%ml summary --dir /content/datalab/tmp/ml/iris/model/summaries
```

Plot loss graphs one based on time and one based on steps.

```python
%ml summary --dir /content/datalab/tmp/ml/iris/model/summaries --name loss
```

Plot loss graph with only steps (no time based graph), and also compare two runs.

Note that the legend of the curves were obtained by "label" file under the specified dir. The label file was created by the training program. If absent, it will use dir full name.

```python
%ml summary --dir /content/datalab/tmp/ml/iris/model/summaries  /content/datalab/tmp/ml/iris/model_lr/summaries --name loss --step
```

Multiple events can be plotted in the same graph too.

```python
%ml summary --dir /content/datalab/tmp/ml/iris/model/summaries  /content/datalab/tmp/ml/iris/model_lr/summaries --name loss queue/batch/fifo_queue/fraction_of_48_full --step
```

Note that the summary dir can be GCS path too.
